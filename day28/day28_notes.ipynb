Step 3: Tech Stack Justification

Why FAISS?

Answer:

"FAISS is fast, open-source vector similarity search library. For prototype and medium-scale documents it provides low latency without external database cost."

Why Sentence Transformer?

Answer:

"It converts text into numerical vectors that represent semantic meaning."

Why LangChain?

Answer:

"LangChain helps connect LLMs, retrievers, prompts and document loaders into one pipeline."
