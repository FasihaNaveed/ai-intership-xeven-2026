# Concept 1: Tuples

A tuple is an ordered and immutable collection in Python.

Immutable means values cannot be changed after creation.

Example:

my_tuple = (10,20,30)

Tuples are faster than lists because they cannot be modified.

They are useful for fixed data like coordinates.

```python
# Creating a tuple

numbers = (10, 20, 30)

print(numbers)

# Accessing tuple values

print(numbers[0])


# Demonstrating immutability

try:
    numbers[0] = 100
except TypeError as error:
    print("Error:", error)
```

# Tuple Packing and Unpacking

Packing:
Multiple values are stored in one tuple.

Unpacking:
Tuple values are assigned into variables.

Example:

a,b = (1,2)

Functions can return multiple values using tuples.

```python
# Tuple packing

student = ("Ali", 90, "AI")

print(student)


# Tuple unpacking

name, marks, field = student

print(name)
print(marks)
print(field)



# Function returning multiple values

def get_student():

    return "Sara", 95


student_name, score = get_student()

print(student_name)
print(score)
```

# Concept 2: Sets

A set is an unordered collection of unique elements.

Features:

- No duplicate values
- Fast lookup
- Mutable collection

```python
# Creating a set

numbers = {1,2,3,3,4,5}


print(numbers)


# Duplicate removed automatically

print(len(numbers))
```

# Set Operations

Union:
Combines all elements.

Intersection:
Common elements.

Difference:
Elements present in first set only.

Symmetric Difference:
Elements that are not common.

```python
set_a = {1,2,3,4}

set_b = {3,4,5,6}


print("Union:")
print(set_a.union(set_b))


print("Intersection:")
print(set_a.intersection(set_b))


print("Difference:")
print(set_a.difference(set_b))


print("Symmetric Difference:")
print(set_a.symmetric_difference(set_b))
```
