# Day 10 Notes: Dictionaries & JSON


## Dictionary

Dictionary stores data in key-value pairs.

Example:

{
"name":"Ali",
"age":20
}


---

## Dictionary Methods

get()
keys()
values()
items()
update()
pop()


---

## Nested Dictionary

Dictionary inside dictionary.

Used for storing complex information.


Example:

student = {

"id":1,

"grades":
{
"Math":90
}

}


---

## JSON

JSON helps save and load data.

json.dump()
writes data.

json.load()
reads data.


---

## Dictionary Comprehension

Creates dictionaries using short syntax.

Example:

{
x:x*x
for x in range(5)
}
