Topic 1: *args

```python
# *args = multiple positional arguments

def add_numbers(*args):
    return sum(args)

print(add_numbers(10, 20))
print(add_numbers(10, 20, 30, 40))
```

*args allows a function to accept any number of positional arguments.

Example:
func(1,2,3,4)

args becomes:
(1,2,3,4)

Topic 2: **kwargs

```python
# **kwargs = multiple keyword arguments

def student_info(**kwargs):
    print(kwargs)

student_info(name="Ali", age=20, city="Lahore")
```

**kwargs allows a function to accept any number of keyword arguments.

Output:
{
'name':'Ali',
'age':20,
'city':'Lahore'
}

Topic 3: Lambda Function

```python
square = lambda x: x*x

print(square(5))
```

Lambda is an anonymous function.

Syntax:
lambda arguments : expression

Topic 4: List Comprehension

```python
numbers = [1,2,3,4,5]

squares = [x*x for x in numbers]

print(squares)
```

List comprehension creates lists in one line.

Syntax:
[expression for item in iterable]

List comprehension creates lists in one line.

Syntax:
[expression for item in iterable]

```python
numbers = [1,2,3,4]

result = {x:x*x for x in numbers}

print(result)
```

Dictionary comprehension creates dictionaries in one line.

Syntax:
{k:v for item in iterable}
