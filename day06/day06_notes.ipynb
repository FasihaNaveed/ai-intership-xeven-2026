1. Lists Basics

```python
my_list = [1, 2, 3, "text", True]
print(my_list)
```

2. Indexing

```python
print(my_list[0])   # first item
print(my_list[-1])  # last item
```

3. Slicing

```python
nums = [1,2,3,4,5,6,7,8]

print(nums[1:4])   # range
print(nums[::2])   # step
```

4. List Methods

```python
names = ["Ali", "Sara", "Ahmed"]

names.append("Usman")
names.insert(1, "Zain")
names.remove("Sara")
```

5. Sorting

```python
names.sort()
names.reverse()
```

Errors Handling Note

Common Error: IndexError
Cause: Wrong index access
Solution: Always check list length before accessing

Comparison Table

| Method    | Use                |
| --------- | ------------------ |
| append()  | Add item at end    |
| insert()  | Add at position    |
| extend()  | Add multiple items |
| remove()  | Delete by value    |
| pop()     | Delete by index    |
| sort()    | Sort list          |
| reverse() | Reverse list       |
