🔷 Task 1: Simple RAG Pipeline

Objective

Build a basic Retrieval-Augmented Generation (RAG) system using FAISS.

🔷 What We Did

1. Document Loading

Text files were loaded from data/ folder.

2. Chunking

Documents were split into smaller chunks so that:

embedding is efficient

search becomes accurate

3. Embedding Generation

Used:

sentence-transformers/all-MiniLM-L6-v2

to convert text → vectors

4. Vector Store Creation

FAISS was used to store embeddings for fast similarity search.

5. Retriever Creation

as_retriever(search_kwargs={"k": 4})

This retrieves top 4 most relevant chunks.

6. Query Execution

User question was converted into embedding and matched with stored vectors.

7. Result

System successfully retrieved correct context from:

rag.txt (most relevant)

plus supporting documents

🧠 Why This is RAG

Because pipeline is:

User Question
   ↓
Embedding
   ↓
FAISS Search
   ↓
Relevant Context
   ↓
(LLM would generate final answer)
