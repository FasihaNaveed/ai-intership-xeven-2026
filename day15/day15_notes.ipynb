🔹 1. What is an LLM?

LLM = Large Language Model

👉 It is an AI that understands and generates text.

Example:
You type: “Explain Python”
AI replies: explanation
🧠 How it works (Simple version)
🔸 Transformer Architecture

LLMs are built on “Transformer model”

👉 It reads text and finds relationships between words.

🔸 Attention Mechanism

👉 Model focuses on important words

Example:

“Ali gave Ahmed his book”

AI understands “his” refers to Ali or Ahmed based on attention.

🔸 Pretraining vs Fine-tuning

| Step        | Meaning                                        |
| ----------- | ---------------------------------------------- |
| Pretraining | Model learns from internet data                |
| Fine-tuning | Model trained for specific task (chat, coding) |

🔹 3. Important Concepts
🧩 Tokens

👉 Text broken into small pieces

Example:

"Hello World"

becomes:

["Hello", "World"]
🧩 Context Window

👉 Memory of AI

Example:

Small window → forgets old chat
Big window → remembers more
🧩 Temperature

👉 Controls creativity

Value	Result

| Value | Result              |
| ----- | ------------------- |
| 0.0   | strict, same answer |
| 0.7   | balanced            |
| 1.5   | creative / random   |

🔹 4. API Parameters

| Parameter         | Meaning              |
| ----------------- | -------------------- |
| max_tokens        | response length      |
| temperature       | creativity           |
| top_p             | variety control      |
| frequency_penalty | repeat words reduce  |
| presence_penalty  | encourage new topics |
