# Day 11 - Loops & Iteration

## For Loop

A for loop is used to iterate through a sequence such as a list, tuple, string, or range.

Example:

```python
for i in range(5):
    print(i)
```

Output:

0
1
2
3
4
```

---

## range()

The `range()` function generates a sequence of numbers.

Example:

```python
range(5)
```

Produces:

```
0, 1, 2, 3, 4
```

---

## enumerate()

The `enumerate()` function returns both the index and the value while iterating through a sequence.

Example:

```python
names = ["Ali", "Sara"]

for index, name in enumerate(names):
    print(index, name)
```

Output:

```
0 Ali
1 Sara
```

---

## zip()

The `zip()` function combines multiple sequences and iterates through them together.

Example:

```python
names = ["Ali", "Sara"]
marks = [80, 90]

for name, mark in zip(names, marks):
    print(name, mark)
```

Output:

```
Ali 80
Sara 90
```

---

## While Loop

A while loop continues executing as long as a condition remains true.

Example:

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

Output:

```
1
2
3
4
5
```

---

## Break

The `break` statement immediately terminates the loop.

Example:

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

Output:

```
0
1
2
3
4
```

---

## Continue

The `continue` statement skips the current iteration and moves to the next iteration.

Example:

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

Output:

```
0
1
3
4
```

---

## Else with Loops

The `else` block executes when a loop finishes normally without encountering a `break` statement.

Example:

```python
for i in range(3):
    print(i)
else:
    print("Finished")
```

Output:

```
0
1
2
Finished
```

---

## Nested Loops

A nested loop is a loop inside another loop. It is commonly used for working with tables, matrices, and two-dimensional data.

Example:

```python
for row in range(3):
    for col in range(3):
        print(row, col)
```

Output:

```
0 0
0 1
0 2
1 0
1 1
1 2
2 0
2 1
2 2
```

### Time Complexity

Nested loops often have a time complexity of:

```text
O(n²)
```

This means that if the input size doubles, the number of operations increases approximately four times.
