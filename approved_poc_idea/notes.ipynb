# 🎯 Mentor Interview Questions (Part 2)

### Q1. Problem Statement kya hai?

Company ki internal knowledge multiple systems mein scattered hoti hai, is wajah se employees ko information dhoondhne mein bohot time lagta hai.

### Q2. Current workflow mein kya problem hai?

Employee pehle documents search karta hai, phir multiple files padhta hai, phir senior se poochta hai. Is process mein time waste hota hai aur knowledge baar-baar repeat hoti hai.

### Q3. Pain Points kya hain?

* Knowledge scattered hai.
* Outdated documents milte hain.
* Experts ko same questions baar-baar answer karne padte hain.
* Analytics nahi hoti.
* Chat mein di hui knowledge baad mein searchable nahi rehti.

### Q4. Permission-aware retrieval kya hota hai?

AI sirf wohi documents retrieve karta hai jinhein dekhne ki user ko permission hoti hai. Agar kisi user ke paas HR documents ka access nahi hai, to AI unse related information return nahi karega.

### Q5. Standard ChatGPT kyun use nahi kar sakte?

Kyun ke uske paas company ke private documents nahi hote. RAG-based assistant har answer company ke actual documents se retrieve karke deta hai.

### Q6. Vector Search aur Keyword Search mein kya difference hai?

Keyword Search exact words ko match karti hai, jab ke Vector Search sentence ka meaning samajh kar relevant documents retrieve karti hai.

### Q2. Modular Architecture kyun use karte hain?
