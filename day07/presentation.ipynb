Slide 4: Best Project Overview

Student Management System

Purpose:
To manage student records using Python lists.

Features:
• Add students
• Insert students at specific positions
• Remove students
• Display student records
• Sort students alphabetically
• List slicing operations

Concepts Used:
• Lists
• List Methods
• Indexing
• Slicing
• Sorting

```python
students = ["Ali", "Ahmed", "Sara", "Ayesha", "Usman"]

students.append("Fatima")

students.insert(1, "Hassan")

students.remove("Ahmed")

students.pop()

students.sort()

print(students)
```

append()  -> adds item at end

insert()  -> adds item at specific position

remove()  -> removes specific item

pop()     -> removes item by index

sort()    -> sorts list alphabetically
