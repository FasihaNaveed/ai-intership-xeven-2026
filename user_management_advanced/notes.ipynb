Project: Advanced User Management API

Features Implemented:

1. Exception Handling using try-except blocks.
2. Logging using Loguru library.
3. Error printing using print() and logger.error().
4. Request and response logging through FastAPI middleware.
5. JSONResponse returned for all APIs.
6. PostgreSQL database integration using SQLAlchemy.
7. pgAdmin used for database management.
8. Delete API modified to use query parameter instead of path parameter.

Status Codes Used:

* 200: Success
* 201: User Created
* 404: User Not Found
* 500: Internal Server Error

Technology Stack:

* FastAPI
* PostgreSQL
* SQLAlchemy
* Loguru
* Python Multipart
