🟦 1. What is a Function?

A function is a reusable block of code that performs a specific task.

Example:
def greet():
    print("Hello World")

greet()
🟦 2. Function with Parameters

Parameters allow us to pass values into a function.

Example:
def greet(name):
    print("Hello", name)

greet("Ali")
🟦 3. Return Statement

Functions can return values using return.

Example:
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
🟦 4. Types of Arguments
✔ Positional Arguments
def info(name, age):
    print(name, age)

info("Ali", 20)
✔ Keyword Arguments
info(age=20, name="Ali")
✔ Default Arguments
def greet(name="Guest"):
    print(name)

greet()
🟦 5. *args (Multiple Arguments)
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)
🟦 6. Scope of Variables
✔ Local Variable
def test():
    x = 10
    print(x)
✔ Global Variable
x = 100

def show():
    print(x)

show()
🟦 7. Global Keyword

Used to modify global variable inside function.

count = 0

def increase():
    global count
    count += 1

increase()
print(count)
🟦 8. Function Design Principles

✔ One function = One task
✔ Clear and meaningful names
✔ Avoid repetition
✔ Use docstrings

🟦 9. Docstrings

Used to describe function purpose.

def add(a, b):
    """
    This function adds two numbers
    """
    return a + b
