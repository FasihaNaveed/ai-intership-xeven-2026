# Day 17: Text Embeddings & Semantic Search


## What are Embeddings?

Embeddings are numerical vector representations of text.

They convert words or sentences into numbers so computers can understand semantic meaning.


Example:

"dog"

becomes:

[0.23,0.45,0.67....]


Similar meanings have similar vectors.


---

## How Embeddings Work

Embedding models use neural networks to encode text.

The model understands relationships between words and places similar texts close together in high dimensional space.


Example:

Dog and Puppy vectors are close.

Dog and Car vectors are far.

## Similarity Metrics


### Cosine Similarity

Measures angle between two vectors.

Value:

1 = very similar

0 = different


Formula:

(A.B)/(||A|| ||B||)



### Euclidean Distance

Measures distance between vectors.


Smaller distance means more similarity.



### Dot Product

Multiplication of vectors.

Higher value means more similarity.

## Neural Network Encoding

Embedding models use neural networks to convert text into numerical vectors.

The model reads the input text and understands relationships between words.

Example:

"dog" and "puppy" have similar meanings, so their vectors are close.


## High Dimensional Space

Embeddings exist in a high-dimensional vector space.

Each dimension represents learned features of text meaning.

Example:

A sentence can be represented as:

[0.23, 0.45, -0.12, 0.78....]

More dimensions allow the model to capture more semantic information.


## Vector Magnitude

||A|| represents the magnitude (length) of a vector.

It is calculated using:

sqrt(a1² + a2² + a3² ...)


In Python:

np.linalg.norm()


## Cosine Similarity Calculation
