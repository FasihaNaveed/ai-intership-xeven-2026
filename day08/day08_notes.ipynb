1. What are Lists in Python?

A List is an ordered collection that can store multiple values in a single variable.

Example

```python
students = ["Ali", "Sara", "Ahmed"]
print(students)
```

Features of Lists
Ordered (position maintain rehti hai)
Mutable (change kar sakte hain)
Multiple data types store kar sakti hain

```python
data = ["Ali", 20, 85.5, True]
```

2. List Methods
append()

List ke end par item add karta hai.

students = ["Ali", "Sara"]
students.append("Ahmed")

print(students)

Output:

```python
['Ali', 'Sara', 'Ahmed']
```

insert()

Specific position par item add karta hai.

```python
students.insert(1, "Ayesha")
```

Output:

```python
['Ali', 'Ayesha', 'Sara', 'Ahmed']
```

remove()

Value remove karta hai.

```python
students.remove("Sara")
```

sort()

Ascending order mein arrange karta hai.

```python
numbers = [5, 2, 9, 1]
numbers.sort()
```

Output:

```python
[1, 2, 5, 9]
```

reverse()

List ulta kar deta hai.

```python
numbers.reverse()
```

clear()

Saari list empty kar deta hai.

```python
numbers.clear()
```

3. List Slicing

Syntax:

```python
list[start:end:step]
```

Example:

```python
numbers = [10,20,30,40,50]

print(numbers[1:4])
```

Output:

```python
[20,30,40]
```

Negative Indexing

```python
numbers[-1]
```

Output:

```python
50
```

Reverse List

```python
numbers[::-1]
```

Output:

```python
[50,40,30,20,10]
```

4. List Comprehensions

Short way to create lists.

Normal way:

```python
squares = []

for x in range(5):
    squares.append(x*x)

print(squares)
```

List comprehension:

```python
squares = [x*x for x in range(5)]

print(squares)
```

Output:

```python
[0,1,4,9,16]
```

Even numbers:

```python
evens = [x for x in range(10) if x % 2 == 0]
```

Output:

```python
[0,2,4,6,8]
```
